```python
img_rgb = imread ("fourier3.jpg")
img_rgb.shape
```

```python
#img_rgb[:,:,0]==img_rgb[:,:,1]
(img_rgb[:,:,0]==img_rgb[:,:,2]).all()
```

```python
img = mean ( img_rgb , axis = -1) # average across the last ( color ) axis
NY , NX = img.shape # The first index , being the array â€™s row , is Y , not X
figure()
imshow ( img , cmap = "gray" )
NX , NY
```

```python
img_fft = fftshift ( fft . fft2 ( img ) )
figure()
imshow ( log ( abs ( img_fft ) ) , cmap = "gray" )
```

```python
xfreq = fftshift ( fftfreq ( NX , d =1.0) )
yfreq = fftshift ( fftfreq ( NY , d =1.0) )
extent_freq = ( min ( xfreq ) , max ( xfreq ), min ( yfreq ) , max ( yfreq ) )
figure()
imshow ( log ( abs ( img_fft ) ) , cmap = "gray" , aspect = NY / NX , extent = extent_freq )
```

```python
almost = fft . ifft2 ( ifftshift ( img_fft ) )
figure ( )
imshow ( almost . real , cmap = "gray" )
```

```python
new_fft = img_fft
new_fft[:,:] = 0
new_fft[NY//2+1,NX//2+3] = 1
new_fft[NY//2+20,NX//2-20] = 0.5
figure ( )
imshow ( new_fft.real , cmap = "gray"  , aspect = NY / NX , extent = extent_freq )
```

```python
almost = fft . ifft2 ( ifftshift ( new_fft ) )
figure ( )
imshow ( almost . real , cmap = "gray" )
```
